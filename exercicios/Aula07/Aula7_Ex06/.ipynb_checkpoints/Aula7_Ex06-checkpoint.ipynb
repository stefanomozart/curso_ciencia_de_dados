### 7.6 Plote o heatmap da correlação das variáveis do dataframe formado pelo merge do dataset de morbidade com o dataset de gini.
#### O merge desses dois datasets for realizado no exercício 6.5
#### Defina um tamanho de figura 35x35, plote o heatmap com anotações e remova a barra de cores com o argumento cbar=False

```python
# Plote seu gráfico aqui

# Plote aqui seu gráfico
# path_morbidade = https://github.com/alexlopespereira/curso_ciencia_dados2020/raw/master/data/originais/morbidade/consolidado/morbidade.csv
# path_gini = 'https://github.com/alexlopespereira/curso_ciencia_dados2020/raw/master/data/originais/gini/ginibr.csv'
path_morbidade = '../../curso_ciencia_dados2020/data/originais/morbidade/consolidado/morbidade.csv'
path_gini = '../../curso_ciencia_dados2020/data/originais/gini/ginibr.csv'


df_morb = pd.read_csv(path_morbidade, sep=';', decimal=',', dtype={'cod_ibge6': str}, na_values=['-', '...'])

df_gini = pd.read_csv(path_gini, sep=';', skiprows=2, skipfooter=2, engine='python', na_values='...', decimal=',')
df_gini['cod_ibge6'] = df_gini['Município'].str[:6]

df_mg = pd.merge(df_morb, df_gini, on='cod_ibge6') 


fig, axc = plt.subplots(figsize=(35,35)) 
sns.heatmap(data=df_mg.corr(), annot=True, cmap='coolwarm', cbar=False, ax=axc)
```
